```python
from pyproj import Transformer 

#WGS 84 / UTM zone 30N , (EPSG:32630)
#WGS 84 / UTM zone 31N , (EPSG:32631)
#WGS 84 / UTM zone 32N , (EPSG:32632)

EPSG = "32630"

transformer_gps_to_lambert = Transformer.from_crs("EPSG:4326", f"EPSG:{EPSG}")
transformer_lambert_to_gps = Transformer.from_crs(f"EPSG:{EPSG}", "EPSG:4326")

if EPSG == "27572":
    center_map_x, center_map_y = (650000, 2080000) # epsg 27572

if EPSG == "27571":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)
    
if EPSG == "27573":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)
    
if EPSG == "27574":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)

if EPSG == "32630":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(45.9, 3.2)

if EPSG == "32631":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)

if EPSG == "32632":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)

    
if EPSG == "27563":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)
    
if EPSG == "27561":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)
    
if EPSG == "27562":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(46, 3)
    
if EPSG == "2154":
    center_map_x, center_map_y = (670000, 6550000) # epsg 2154
    
if EPSG == "3857":
    center_map_x, center_map_y = transformer_gps_to_lambert.transform(47, 2) # epsg 3857


position_station_x, position_station_y = transformer_gps_to_lambert.transform(position_noyal[0], position_noyal[1])
print(position_station_x, position_station_y)

position_aja_station_x, position_aja_station_y = transformer_gps_to_lambert.transform(position_ajacio[0], position_ajacio[1])
print(position_aja_station_x, position_aja_station_y)

position_ale_station_x, position_ale_station_y = transformer_gps_to_lambert.transform(position_aleria[0], position_aleria[1])
print(position_ale_station_x, position_ale_station_y)
```
